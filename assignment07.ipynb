# import numpy&matpyplot

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

# plotting the noisy data (x,y1)

```python
num     = 300
std     = 5 
def fun(x): 
    f = np.abs(x) * np.sin(x)
    return f

n       = np.random.rand(num)
nn     = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std
plt.subplots()
plt.plot(x, y1, 'k.')
plt.show()
```

# plotting the clean data (x,y2)

# We define a model $\hat{f}(x)=θ_0 x^0 +θ_1 x^1 +···+θ_p x^p $.
# First, we define a matrix creating function matA.

```python
y2      = y1 + nn * std
pt2    =plt.subplots()
plt.plot(x, y2, 'k.')
plt.show()
```

```python
def matA(x,p):
    A=np.zeros((len(x),p))
    for i in range(len(x)):
        for j in range(p):
            A[i,j]=x[i]**j
    return A
```

# Next we define a function to get model parameters.

```python
def mp(x,p,y2):
    A=matA(x,p)
    A_t=A.transpose()
    gram_A=np.matmul(A_t,A)
    gram_A_i=inv(gram_A)
    ps_A=np.matmul(gram_A_i,A_t)
    mp=np.matmul(ps_A,y2)
    return mp
```

# Finally, we can define a polynomial fitting function, poly_fit(). A pf_x() function computes all values along the x-axis. 

```python
def poly_fit(a,x,y2,p):
    m_p=mp(x,p+1,y2)
    f=0
    for i in range(p):
        f=f+m_p[i]*(a**i)
    return f
def pf_x(x,y2,p):
    f0=np.zeros(num)
    for i in range(len(x)):
        f0[i]=poly_fit(x[i],x,y2,p)
    return f0
```

# We plot polynomial fitting function for p from 0 to 9.

```python
for p in range(10):
    f=pf_x(x,y2,p+1)
    pt3=plt.subplots()
    plt.plot(x,f,'k.',x,y2,'b.')
    plt.show()
```

# A residual of polynomial fitting function for each p's.

```python
def residual(x,y):
    r=0
    for i in range(len(x/1.5)):
        r=r+(y[i]-x[i])**2
        return r
for k in range(10):
    print("p=%d:\t\t%f\n"%(k,residual(pf_x(x,y2,k+1),y2)))
```
