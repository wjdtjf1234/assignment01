```python
# 함수 연산과 그래프를 그리기 위한 API들을 가져옵니다.
import numpy as np
import matplotlib.pyplot as plt
```

```python
#myFunction함수는 우리가 Taylor Approximation을 시킬 함수입니다. 저는 $\x^2$ 함수를 사용하기로 하였으며, 다음과 같이 정의합니다.

def myFunction(x):
    f= x**2
    return f
```

```python
# myDerivativeFunction은 myFunction을 미분한 함수로써 Taylor Approximation에서 사용되므로 이 또한 다음과 같이 정의해줍니다. 
def myDerivativeFunction(x):
    Df = x*2
    return Df
```

```python
# x 축을 np.arange 함수를 사용하여 범위를 정해 줍니다. 첫 번째 인자는 시작점, 두 번째 인자는 끝나는 점, 마지막 인자는 스텝 단위입니다.
x = np.arange(-15, 15, 0.1)
```

```python
# f, Df로 함수 객체를 선언한 후, plt.subplots함수를 사용하여 함수의 그래프를 출력할 ax를 선언한 뒤, f는 파란색, Df는 빨간 색으로 그려줍니다.
f = myFunction(x)
Df = myDerivativeFunction(x)
fig, ax= plt.subplots()
ax.plot(x,f,'b')
ax.plot(x,Df,'r')
plt.show()
```

```python
# 결과는 위와 같으며, 위에서 정의한 $f= \x^2$와 $Df= \2*x$가 그래프로 출력된 것을 알 수 있습니다.
```

```python
# 다음은 위에서 정의한 함수에 Taylor Approximation을 적용시켜 볼 차례입니다. 먼저 3개의 점을 택한 뒤, 해당 좌표에서 그려지는 
# Taylor Approximation을 관찰합니다. myFunction에서 정의한 함수의 Taylor Approximation은 다음과 같습니다.
def TaylorApproximation(x,a):
    Tf= a**2 + 2*a*(x-a)
    return Tf
```
